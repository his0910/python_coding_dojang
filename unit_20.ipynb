## UNIT 20 FizzBuzz 문제

매우 간단한 프로그래밍 문제이며, 프로그래밍 면접에 자주 등장하는 문제.
문제 조건을 꼼꼼히 따지지 않으면 경력자도 실수하기 쉽고, 면접에서 탈락하기도 함.
그만큼 기초 실력을 가늠하는 문제이기 때문에 잘 알아두면 좋음.

### 20.1 1부터 100까지 숫자 출력하기

```python
for i in range(1, 101):
    print(i)
```

### 20.2 3의 배수일 때와 5의 배수일 때 처리하기

```python
for i in range(1, 101):
    if i % 3 == 0:
        print('Fizz')
        
    if i % 5 == 0:
        print('Buzz')
        
    else:
        print(i)
```

### 20.3 3과 5의 공배수 처리하기

```python
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print('FizzBuzz')
        
    if i % 3 == 0:
        print('Fizz')
        
    if i % 5 == 0:
        print('Buzz')
        
    else:
        print(i)
```

### 20.4 논리 연산자를 사용하지 않고 3과 5의 공배수 처리하기

```python
for i in range(1, 101):
    if i % 15 == 0:
        print('FizzBuzz')
        
    if i % 3 == 0:
        print('Fizz')
        
    if i % 5 == 0:
        print('Buzz')
        
    else:
        print(i)
```

### 20.5 코드 단축하기

```python
for i in range(1, 101):
    print('Fizz' * (i%3 == 0) + 'Buzz' * (i%5 == 0) or i)
# 문자열 곱셈과 덧셈을 이용하여 print 안에서 처리
```

### 20.7 연습문제 : 2와 11의 배수, 공배수 처리하기

다음 소스 코드를 완성하여 1부터 100까지의 숫자를 출력하면서 2의 배수일 때는 'Fizz', 11의 배수일 때는 'Buzz', 2와 11의 공배수일 때는 'FizzBuzz'가 출력되게 만드세요.

```python
for i in range(1, 101):
    if i % 2 == 0 and i % 11 == 0:
        print('FizzBuzz')
        
    elif i % 2 == 0:
        print('Fizz')
        
    elif i % 11 == 0:
        print('Buzz')
        
    else:
        print(i)
```

### 20.8 심사문제 : 5와 7의 배수, 공배수 처리하기

표준 입력으로 정수 두 개가 입력됩니다(첫 번째 입력 값의 범위는 1 ~ 1000, 두 번째 입력 값의 범위는 10 ~ 1000이며 첫 번째 입력 값의 범위는 두 번째 입력 값보다 항상 작습니다). 첫 번째 정수부터 두 번째 정수까지 숫자를 출력하면서 5의 배수일 때는 'Fizz', 7의 배수일 때는 'Buzz', 5와 7의 공배수일 때는 'FizzBuzz'를 출력하는 프로그램을 만드세요(input에서 안내 문자열은 출력하지 않아야 합니다).

```python
a, b = map(int, input().split())

for i in range(a, b+1):
    if i%5 == 0 and i%7 == 0:
        print('FizzBuzz')
    
    elif i%5 == 0:
        print('Fizz')
        
    elif i%7 == 0:
        print('Buzz')
        
    else:
        print(i)
```
