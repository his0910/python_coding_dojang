## UNIT 17 while 반복문으로 Hello, world! 100번 출력하기

### while 반복문
조건식을 판별하면 True이면 코드를 계속 반복, False이면 반복문을 끝낸 뒤 다음 코드를 실행

### 17.1 while 반복문 사용하기

```python
i = 0
while i < 100:
    print('Hello, world!')
    i += 1
```

```python
# 초깃값을 감소시키기
i = 100
while i > 0:
    print('Hello, world!', i)
    i -= 1
```

```python
# 입력한 횟수대로 반복하기
count = int(input('반복할 횟수를 입력하세요 : '))

i = 0
while i < count: # i가 count보다 작을 때 반복
    print('Hello, world!', i)
    i += 1
```

```python
random.randint(1, 6) # 주사위
```

```python
import random

i = 0
while i != 3: # 3이 아닐 때 계속 반복
    i = random.randint(1, 6) # randint를 사용하여 1과 6 사이의 난수를 생성
    print(i)
```

```python
# random.choice 함수
dice = [ 1, 2, 3, 4, 5, 6 ]
random.choice(dice)

# 리스트뿐만 아니라 튜플, range, 문자열도 가능
```

### 17.5 연습문제 : 변수 두 개를 다르게 반복하기

다음 소스 코드를 완성하여 정수 2 5, 4 4, 8 3, 16 2, 32 1이 각 줄에 출력되게 만드세요. while에 조건식은 두 개 지정하고, 두 변수를 모두 변화시켜야 합니다.

```python
i = 2
j = 5

while i <= 32 or j >= 1:
    print(i, j)
    i *= 2
    j -= 1
```

### 17.6 심사문제 : 교통카드 잔액 출력하기

표준 입력으로 금액(정수)이 출력됩니다. 1회당 요금은 1,350원으로 교통카드를 사용할 때마다 각 줄에 잔액을 출력하는 프로그램을 만드세요(input에서 안내 문자열은 출력하지 않아야 합니다). 단, 최초 금액은 출력하지 않아야 합니다. 그리고 잔액은 음수가 될 수 없으며 잔액이 부족하면 출력을 끝냅니다.

```python
x = int(input())

while x >= 1350:
    x -= 1350
    print(x)
```
